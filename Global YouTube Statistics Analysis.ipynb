### Data Cleaning

```python
#Removing unnecessary columns 
```

```python
df2 = df[['rank', 'Youtuber', 'subscribers', 'video views', 'category', 
     #'Title',
     #'uploads', 
     'Country', 
     #'Abbreviation', 'channel_type',
     #'video_views_rank', 'country_rank', 'channel_type_rank',
       'video_views_for_the_last_30_days', 
     #'lowest_monthly_earnings',
     #'highest_monthly_earnings', 'lowest_yearly_earnings',
     #'highest_yearly_earnings', 
     'subscribers_for_last_30_days',
       'created_year', 'created_month', 
     #'created_date',
     #'Gross tertiary education enrollment (%)', 
     'Population',
     #'Unemployment rate', 'Urban_population', 'Latitude', 'Longitude'
    ]].copy()
```

```python
#Changing column names to uppercase 
```

```python
df3 = df2.rename(columns = {"rank":"Rank",
                    'subscribers':"Subscribers",
                    'video views': 'Video_Views',
                    'category':'Category',
                    'video_views_for_the_last_30_days': 'Video_Views_for_the_Last_30_Days',
                    'subscribers_for_last_30_days': 'Subscribers_for_Last_30_Days',
                    'created_year':'Created_Year',
                    'created_month' : 'Created_Month'})
```

### Feature Understanding 

```python
#Understanding the distribution of the year when the Youtube videos were published
```

```python
df3['Created_Year'].value_counts().head(5)
```

```python
#Visualisng distibution of when Youtube videos wered published  
```

```python
ax = df3['Created_Year'].plot(kind='hist',range=(2001, 2022)
                              ,bins=22,
                              title = 'Disribution of Year of Video Publication')

ax.set_xlabel('Year of Publication')            
plt.show()
```

```python
# Create top-5 dataset for visualization

```

```python
df_top5 = df3[df3["Rank"]<=5]
print(df_top5)
```

```python
#Identify the most popular categories on Youtube by Subscribers 
```

```python
plt.bar(df_top5.Category,df_top5.Subscribers, color='green')
plt.title('Top 5 Video Categories by Number of Subscribers')
plt.xlabel('Video Category')
plt.ylabel('Number of Subscribers')

plt.xticks(rotation=90)
```

```python
#Identifying the top 5 Youtubers by subscribers 
```

```python
plt.bar(df_top5.Youtuber,df_top5.Subscribers)
plt.xlabel('Youtuber')
plt.ylabel('Number of Subscribers')
plt.title('Top 5 Youtubers by Subscribers')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.plot(df_top5.Youtuber,df_top5.Video_Views)
plt.ylabel ('Video Views')
plt.title('Video Views by Top 5 Youtubers')
plt.xticks(rotation=90)
plt.show()
```

### Feature Relationship

```python
#Exploring the relationship between the number of subscribers and video views with a scatterplot
```

```python
plt.scatter(df3.Subscribers, df3.Video_Views)
plt.xlabel('Subscribers')
plt.ylabel('View Views')
plt.title('Video Views vs Subscribers')

plt.ticklabel_format(style="plain")

plt.show()
```

```python
#Getting a yearly breakdown of the corelation between subscribers and video views 
```

```python
ax = sns.scatterplot(x='Subscribers',
                y='Video_Views',
                hue='Created_Year',
                data=df3).set(title='Corelation Between Video Views and Subscribers, with Year of Creation Information')
plt.show()
```

```python
#Boxplot showing the distribution of videos viedeos by the month videos were created 
```

```python
sns.boxplot(data=df3, x="Video_Views", y="Created_Month").set(title='Distribution of Video Views by Month')
```

```python
#Corelation between numeric values  
```

```python
df_corr = df[['rank','video views','subscribers','highest_monthly_earnings']].dropna().corr()
df_corr
```

```python
#Visulisating the corelations with a heatmap
```

```python
sns.heatmap(df_corr, annot=True)
plt.title('Corelation Between Numeric Values')
plt.show()
```

```python
#Showcasing how each of the nummeric elements from the dataset interact with each other 
```

```python
sns.pairplot(df,
             vars=['rank','video views','subscribers','highest_monthly_earnings'], hue = 'created_year')
plt.show()
        
```
